```python
!pip install qrcode[pil]
import qrcode

# Data for encoding
data = "Hi thanks for taking time to scan my qr code. Have a great day "

# Create QR code instance
qr = qrcode.QRCode(
    version=1,
    box_size=10,
    border=5
)

# Add data
qr.add_data(data)
qr.make(fit=True)

# Create image
img = qr.make_image(fill_color="black", back_color="white")

# Save the image
img.save("qrcode.png")
```
